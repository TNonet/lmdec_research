```python
fig2 = plt.figure(figsize=(10,7))

color_dict = {'160K_640K':'red', '120K_480K':'blue'}
for matrix in ['160K_640K', '120K_480K']:
    work_list = []
    step_list = []
    for work, step in df[(df['Matrix'] == matrix)][['worker','step']].values:
        work_list.append(work)
        step_list.append(step)
    plt.scatter(work_list, [np.median(step) for step in step_list], c=color_dict[matrix], label=matrix)
        
        
plt.title('Iteration runtime vs CPUs')
plt.legend()
plt.ylabel('Seconds')
plt.xlabel('CPUs')
plt.grid()
plt.show()
```
